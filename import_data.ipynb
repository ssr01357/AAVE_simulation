```python
import pandas as pd
import json
import os
```

```python
for model in ["gpt4o", "llama", "mixtral"]:
    result_dir = f"/home/yl3427/cylab/AAVE/vllm/output_{model}"
    import_dir = f"/secure/shared_data/studio_label/AAVE_{model}"
    os.makedirs(import_dir, exist_ok=True)

    for filename in os.listdir(result_dir):
        result_file_path = os.path.join(result_dir, filename)
        with open(result_file_path, "r", encoding="utf-8") as file:
            content = file.read()
        pieces = content.split("||")
        pieces = [piece.strip() for piece in pieces]

        data = {"data": {"dialogues": []}}
        for i, piece in enumerate(pieces, 1):
            data["data"]["dialogues"].append({"author": f"answer {i}", "text": piece})

        import_file_path = os.path.join(import_dir, filename.replace(".txt", ".json"))
        with open(import_file_path, "w", encoding="utf-8") as json_file:
            json.dump(data, json_file, indent=4, ensure_ascii=False)
        print(f"Processed {filename} -> {import_file_path}")
```
